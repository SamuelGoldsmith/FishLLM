```python
from google import genai

def askLLM(prompt, meta = "", context = ""):
    client = genai.Client()
    response = client.models.generate_content(
    model="gemini-3-flash-preview", contents=f'''
    ${meta}
    ${prompt}
    ${context}
    '''
)
    return response.text
```

```python
def getWeather():
    return "Worcester: 67degrees and sunny with a chance of rain at noon"
```

```python
functions = [{"name": 'getWeather', "description":"recieves weather in worcester", "function": getWeather},]
def getMCPContext(prompt):
    query = f'''
    Would query: ${prompt} need context from the following functions:
    ["name": 'getWeather', "description":"recieves weather in worcester"]
    Only respond with the function name or 'none'
    '''
    fun = askLLM(query)
    print(fun)
    if fun == 'none': return ''
    for f in functions:
        if f['name'] == fun: return f['function']()
```

```python


prompt = "What is the weather at WPI today?"


context = getMCPContext(prompt)
if context:
    context = "use the context " + context
print(context)

meta= "this response will be spoken: "
askLLM(meta, context, prompt)
```
